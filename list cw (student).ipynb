````python
# Ref: How to read csv
# DO NOT use other libraries eg: pandas
import csv

with open('dse.csv', encoding='utf-8-sig') as f:
    reader = csv.reader(f)
    data = list(reader)

print(data)
````

````python
# Q1
# -----------------------Type your code-----------------------
print(data[2])
````

````python
# Q2
# -----------------------Type your code-----------------------
print(data[0][3] + ":",data[2][3])
for i in range(1, 8):
    print(data[0][3 + i] + ":", data[2][3 + i])

````

````python
# Q3
# -----------------------Type your code-----------------------
sum = 0
newValue = [0]
newList = []
newValue[0] = int(data[2][3])
newList = newList + newValue
for i in range(1, 7):
    newValue[0] = (int(data[2][3 + i]) - int(data[2][2 + i]))
    newList = newList + newValue
newValue[0] = int(data[2][10])
newList = newList + newValue
for i in range(len(newList)):
    sum = sum + newList[i]




print("list:", newList)
print("Sum of the list:", sum)
````
